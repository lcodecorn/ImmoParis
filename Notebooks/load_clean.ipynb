```python
import pandas as pd
import os

notebook_dir = os.getcwd()
project_root = os.path.dirname(notebook_dir)
csv_path21 = os.path.join(project_root, "Data", "75_2021.csv")
csv_path22 = os.path.join(project_root, "Data", "75_2022.csv")
csv_path23 = os.path.join(project_root, "Data", "75_2023.csv")
csv_path24 = os.path.join(project_root, "Data", "75_2024.csv")
csv_path25 = os.path.join(project_root, "Data", "75_2025.csv")

df21 = pd.read_csv(csv_path21)
df22 = pd.read_csv(csv_path22)
df23 = pd.read_csv(csv_path23)
df24 = pd.read_csv(csv_path24)
df25 = pd.read_csv(csv_path25)
```

```python
df = pd.concat([df21, df22, df23, df24, df25], axis=0)
df.head()
```

```python
pd.set_option('display.max_columns', 40)

df.describe(include="all").T
```

```python
df.isna().sum().sum()
```

```python
col_to_keep = ["id_mutation", "date_mutation", "nature_mutation", "valeur_fonciere", "adresse_nom_voie", "code_postal",
"type_local", "surface_reelle_bati", "nombre_pieces_principales", "longitude", "latitude"]

df = df[col_to_keep]

df.head()
```

```python

df = df[df.type_local == "Appartement"]
df = df[df.nature_mutation == "Vente"]
df = df[df.surface_reelle_bati > 9]
df['nombre_pieces_principales'] = df['nombre_pieces_principales'].replace(0, 1)


df['valeur_fonciere'] = df.groupby('code_postal')['valeur_fonciere'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df["price_per_sqrtm"] = df["valeur_fonciere"] / df["surface_reelle_bati"]

df.head()
```

```python
Q1 = df['price_per_sqrtm'].quantile(0.25)
Q3 = df['price_per_sqrtm'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1 * IQR

df = df[(df['price_per_sqrtm'] >= lower) & (df['price_per_sqrtm'] <= upper)]
```

```python
df = df.drop(columns=["type_local", "nature_mutation"])
```

```python
df.to_csv("../Data/xp.csv")
```
